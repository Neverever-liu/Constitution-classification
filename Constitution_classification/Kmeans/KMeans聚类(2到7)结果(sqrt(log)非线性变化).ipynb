```python
from sklearn import metrics
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import davies_bouldin_score
from sklearn.metrics import calinski_harabasz_score
from sklearn.cluster import KMeans
plt.rcParams['font.sans-serif'] = ['Microsoft YaHei']


data = pd.read_csv('E:/databag/bag2/KMeans聚类(2到7)结果(sqrt(log)非线性变化).csv',index_col=0)
data1=data.drop('4个簇',axis=1)
data1
data2=data1.drop('5个簇',axis=1)
data3=data2.drop('6个簇',axis=1)
data4=data3.drop('7个簇',axis=1)
data4
data5=data4.drop('3个簇',axis=1)
data6=data5.drop('2个簇',axis=1)
```

```python
#存放设置不同簇数时的SSE值
sse_list =[]
# 轮廓系数
silhouettes = []
#Calinski-Harabaz Index
scores = []
#Davies-Bouldin Index
davies_list = []
# 循环设置不同的聚类簇数
for i in range(4, 10):
    y_pre = np.array(data.iloc[:,23+i])
    model = KMeans(n_clusters=i)
    model.fit(data4)
    # kmeans算法inertia属性获取簇内的SSE
    sse_list.append(model.inertia_)
    # 轮廓系数
    silhouette = metrics.silhouette_score(data4, y_pre, metric='euclidean')
    silhouettes.append(silhouette)
    #Davies-Bouldin Index
    davies_list.append(davies_bouldin_score(data4,y_pre))
    #Calinski-Harabaz Index
    score = calinski_harabasz_score(data4,y_pre)
    scores.append(score)
```

```python
# 绘制簇内误方差曲线
plt.subplot(421)
plt.title('簇内误方差')
plt.plot(range(2, 8), sse_list, marker='*')
plt.xlabel('簇数量')
plt.ylabel('簇内误方差(SSE)')
# 绘制轮廓系数曲线
plt.subplot(422)
plt.title('轮廓系数')
plt.plot(range(2, 8), silhouettes, marker='o')
plt.xlabel('簇数量')
plt.ylabel('轮廓系数')
# 绘制Calinski-Harabaz Index曲线
plt.subplot(423)
plt.title('Calinski-Harabaz Index')
plt.plot(range(2, 8),scores , marker='o')
plt.xlabel('簇数量')
plt.ylabel('Calinski-Harabaz Index')

# 绘制Davies-Bouldin Index曲线
plt.subplot(424)
plt.title('Davies-Bouldin Index')
plt.plot(range(2, 8),davies_list , marker='o')
plt.xlabel('簇数量')
plt.ylabel('Davies-Bouldin Index')

plt.suptitle('KMeans聚类(2到7)结果(sqrt(log)非线性变化)')
plt.tight_layout()
plt.show()
```
